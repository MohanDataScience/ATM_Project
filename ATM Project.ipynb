# ATM Project Given By Codingrad

```python
import datetime
import sys
class Atm:
    def __init__(self,name,balance=0):
        self.name=name
        self.balance=balance
    
    def diposite(self,amount):
        
        self.balance=self.balance+amount
        print("-"*70,"\n After Diposite Your Balance:",self.balance," \n","-"*70)
        
    
    def withdrawl(self,amount):
        if self.balance>=amount:
            self.balance=self.balance-amount
            print("-"*70,"\n After Withdrawl Your Balance: ",self.balance,"\n","-"*70)

        else:
            print("\n\n\nINSUFICIENT FUNDS!!!\n\n\n")
    
    def mini_statement(self):
        for i in range(len(statement)):
            
            print(time[i],"---",statement[i])
        print("-"*70,"\n\nnet balance: ",self.balance,"\n","-"*70)
    def bal(self):
        print("-"*70,"\nYour Available Balance:",self.balance,"\n","-"*70)

        
        

print("*WELCOME TO HDFC BANK*")
name=input("Please insert your card for the banking:")
obj=Atm(name)
pin=int(input("ENTER YUOR PIN:"))
if pin == 1234:
    statement=[]
    time=[]
    while True:
        print(""" Dear Customer Please How can I help You:
                        d-Diposite
                        w-Withdrawl
                        b-Balance
                        m-Mini Statement
                        e-Exit             """)
    
        option=input("Choose your Option: ")
        if option=="d" or option=="D":
            amount=int(input("Please Enter your amount to Diposite: "))
            if amount%100==0:
                obj.diposite(amount)
                statement.append(amount)
                time.append(datetime.datetime.now())
            else:
                print('''allowed currency only any number of 
                                                100 rupee or 
                                                500 rupee or 
                                                2000 rupee notes only''')
        elif option=="w" or option== "W":
            amount=int(input("Please Enter your amount to Withdrawl: "))
            if amount%100==0:
                obj.withdrawl(amount)
                statement.append(-amount)
                time.append(datetime.datetime.now())
                
            else:
                print("please enter rupees only divisible by 100")
        elif option=="b" or option=="B":
            obj.bal()
        elif option=="m" or option=="M":
            obj.mini_statement()
        elif option=="e" or option=="E":
            sys.exit()
              
else:
    print("INCORRECT PIN")
```
